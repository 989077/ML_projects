```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_1 = np.array(['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','Nov','Dec'])
data_2 = np.array([100,250,350,45,550,650,70,890,950,1500,1510,1250])
```

```python
# plot line plot
plt.figure(figsize=(6,4))
# plot showing sales trends over the year
plt.plot(data_1,data_2,marker='.',linestyle='-',color='b')
# line plot showing sales trends over the year
plt.title('monthly sales data')
# label for the x-axis
plt.xlabel('Month')
# label for the y-axis
plt.ylabel('sales')
# show grid lines for better readablity
plt.grid(True)
# Display the plot
plt.show()
```

```python
# Histrogram

# create a dataset for histogram
np.random.seed(0) # for reproducibility
data = np.random.normal(loc=0, scale = 1 , size =1000 )
plt.figure(figsize=(6,4))
# histrogram showing the distibution of data
plt.hist(data, bins = 30, color='purple', edgecolor='black')
plt.title('random data Distributiion') # Histrogram of data distibution
plt.xlabel('value')
plt.ylabel('frequency')
plt.grid(True)
```

```python
# Scatter plot
# example data for scatter plot
np.random.seed(1) # random x values
x = np.random.rand(80) * 100
y = np.random.rand(80) * 100

plt.figure(figsize=(6,4))
plt.scatter(x,y, color='red', alpha=0.7) # scatter plot showing relation between x and y
plt.title('random scatter plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()

```

```python
# Bar plot
# example data for bar plot
categories = np.array(['A','B','C','D','E'])
values = np.array([10,20,30,40,50])

plt.figure(figsize=(6,4))
plt.bar(categories, values, color=['blue','green','red','purple','orange'])
plt.title('category values')
plt.xlabel('category')
plt.ylabel('value')
plt.grid(False)
plt.show()
```

```python
# Box plot
np.random.seed(2)
group1 = np.random.normal(loc=0, scale=1, size=100)
group2 = np.random.normal(loc=0, scale=1.5, size=100)

plt.figure(figsize=(6, 4))
plt.boxplot([group1, group2], labels= ['group1','group2'])
plt.grid(True)
```

```python
# Heatmap (Correction Matrix)
# Create a correction matrix for deemostration
np.random.seed(3)
data_matrix = np.random.rand(10,10)

plt.figure(figsize=(6,4))
plt.imshow(data_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar(label = 'Value')
plt.title('Heat map')
```

```python
# Generating some simple data using arrays
x = np.linspace(0,10,100) # 100 values from 0 to 10
y1 = np.sin(x) # sine wave
y2 = np.cos(x)

#Create a figure
plt.figure(figsize=(6,2))

# First subplot: size wave 
plt.subplot(1,2,1) # 1 row, 2 columns, first plot
plt.plot(x,y1,'b--') # line plot for sine wave
plt.title('Sine wave')

# second subplot
plt.subplot(1,2,2) # 1 row, 2 columns, second plot
plt.plot(x, y1, 'r--') # line plot for sine wave
plt.title('coSine wave')

# adjust layout
plt.tight_layout()

# Displaythe plots
plt.show()
```

```python
# Import neccessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
data = pd.read_csv('customers-10000.csv')
print(data.head())
```

```python
# Set the style for plots
sns.set(style='whitegrid') # Use a while grid style for better visualisation
```

```python
# Histogram with KDE

plt.figure(figsize=(6,4))
sns.histplot(data['City'], bins=20, kde=True, color='purple')
plt.title('Distribution of spending Score')
plt.xlabel('Spending Score')
plt.ylabel('Frequency')
plt.show()
```
